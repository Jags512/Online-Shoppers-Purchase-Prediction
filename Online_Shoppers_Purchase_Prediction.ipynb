```python

datashow[['VisitorType','Weekend','Revenue']].value_counts()
X=datashow.drop('Revenue',axis=1)
y=datashow['Revenue']
```

```python
X.shape
```

```python
y.shape
```

```python

from sklearn.preprocessing import OrdinalEncoder   # for muticlass
from sklearn.preprocessing import LabelEncoder     # for single class

encoder = OrdinalEncoder()
X[['VisitorType', 'Weekend','Month' ]] = encoder.fit_transform(
    X[['VisitorType', 'Weekend','Month']]

)
label=LabelEncoder()
y=label.fit_transform(y)


```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=0, stratify=y
)
```

```python
X_train.shape
```

```python
y_train.shape
```

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
X_train=scaler.fit_transform(X_train)
X_test=scaler.transform(X_test)
```

```python
from sklearn.linear_model import LogisticRegression
model=LogisticRegression()
model.fit(X_train,y_train)
model.score(X_test,y_test)
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix
y_pred = model.predict(X_test)
print("Accuracy:", accuracy_score(y_test, y_pred))
print("Confusion Matrix:\n", confusion_matrix(y_test, y_pred))
```
